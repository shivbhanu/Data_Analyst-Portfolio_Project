## Data Insights

```python
# Locations with fastest roller coasters(min of 10 coasters at that location)?
```

```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"') \
    .groupby(['Location'])['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind = 'barh', title = 'Avg Coaster Speed by Location')

ax.set_xlabel('Avg_Coaster_Speed')

plt.show()
```
